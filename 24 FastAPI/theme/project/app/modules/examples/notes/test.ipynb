```python
# app/modules/examples/notes/test.ipynb

# Импорт зависимостей
import requests
import json

# Базовая конфигурация
base_url = "http://127.0.0.1:8000/examples/notes"  
headers = {"Content-Type": "application/json"}
```

```python
# POST — создание заметки
payload = {
    "title": "Моя первая заметка",
    "content": "Это пример текста заметки"
}
resp = requests.post(base_url, headers=headers, data=json.dumps(payload))
print(resp.status_code)
print(resp.json())
note_id = resp.json().get("id")  # сохраняем id для дальнейших тестов
```

```python
# GET — получить все заметки
resp = requests.get(base_url, headers=headers)
print(resp.status_code)
print(resp.json())
```

```python
# GET — получить заметку по ID
resp = requests.get(f"{base_url}/{note_id}", headers=headers)
print(resp.status_code)
print(resp.json())
```

```python
# PATCH — частично обновить заметку
patch_payload = {
    "title": "Обновлённая заметка"
}
resp = requests.patch(f"{base_url}/{note_id}", headers=headers, data=json.dumps(patch_payload))
print(resp.status_code)
print(resp.json())
```

```python
# PUT — полностью заменить заметку
put_payload = {
    "title": "Полностью новая заметка",
    "content": "Новый контент"
}
resp = requests.put(f"{base_url}/{note_id}", headers=headers, data=json.dumps(put_payload))
print(resp.status_code)
print(resp.json())
```

```python
# DELETE — удалить заметку
resp = requests.delete(f"{base_url}/{note_id}", headers=headers)
print(resp.status_code)
```

```python
# GET после удаления (проверка)
resp = requests.get(f"{base_url}/{note_id}", headers=headers)
print(resp.status_code)
print(resp.json() if resp.status_code == 200 else "Заметка удалена")
```
