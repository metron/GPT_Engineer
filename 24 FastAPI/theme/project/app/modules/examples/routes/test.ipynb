```python
# app/modules/examples/routes/test.ipynb

# Импорт зависимостей
import requests
import json

# Базовая конфигурация
base_url = "http://127.0.0.1:8000/examples/routes"  
headers = {"Content-Type": "application/json"}
```

```python
# GET /hello — простой маршрут без параметров

resp = requests.get(f"{base_url}/hello")
print(f"Статус: {resp.status_code}")
print(f"Ответ: {resp.json()}")
```

```python
# GET /square/{number} — параметр пути

resp = requests.get(f"{base_url}/square/7")
print(f"Статус: {resp.status_code}")
print(f"Ответ: {resp.json()}")
```

```python
# GET /search — query‑параметры

params = {
    "q": "test",
    "limit": 5
}
resp = requests.get(f"{base_url}/search", params=params)
print(f"Статус: {resp.status_code}")
print(f"URL запроса: {resp.url}")
print(f"Ответ: {resp.json()}")
```

```python
# POST / — создание ресурса

data = {
    "name": "Иван Петров",
    "birthday": "15.10.2001",
    "admin": True
}
resp = requests.post(
    f"{base_url}",
    headers=headers,
    data=json.dumps(data)
)
result = resp.json()
item_id = result['data']['id']
print(f"Статус: {resp.status_code}")
print(f"Ответ: {result}")
```

```python
# GET /{item_id} — получение ресурса по ID

item_id = 2
resp = requests.get(f"{base_url}/{item_id}")
print(f"Статус: {resp.status_code}")
print(f"Ответ: {resp.json()}")
```

```python
# PUT /{item_id} — обновление ресурса

data = {
    "name": "Иван Владимирович Петров"
}
resp = requests.put(
    f"{base_url}/{item_id}",
    headers=headers,
    data=json.dumps(data)
)
print(f"Статус: {resp.status_code}")
print(f"Ответ: {resp.json()}")
```

```python
# DELETE /{item_id} — удаление ресурса

resp = requests.delete(f"{base_url}/{item_id}")
print(f"Статус: {resp.status_code}")
if resp.status_code == 204:
    print("Ресурс успешно удалён (ответ без тела)")
else:
    print(f"Ответ: {resp.json()}")
```

```python
# DELETE /{item_id} — удаление несуществующего ресурса

resp = requests.delete(f"{base_url}/{item_id}")
print(f"Статус: {resp.status_code}")
if resp.status_code == 204:
    print("Ресурс успешно удалён (ответ без тела)")
else:
    print(f"Ответ: {resp.json()}")
```
