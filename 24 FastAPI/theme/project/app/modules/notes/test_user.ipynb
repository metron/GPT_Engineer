```python
# app/modules/notes/test_user.ipynb

# Инициализация и общие настройки
import requests
import json

BASE_URL = "http://127.0.0.1:8000"
AUTH_URL = f"{BASE_URL}/auth"
NOTES_URL = f"{BASE_URL}/notes"

HEADERS_JSON = {
    "Content-Type": "application/json"
}

def print_response(resp):
    print("Status:", resp.status_code)
    try:
        data = resp.json()
        print("Response:", data)
        if isinstance(data, dict) and "detail" in data:
            print("Detail:", data["detail"])
    except Exception:
        print("Response text:", resp.text)
```

```python
# POST /auth/register — регистрация нового пользователя

payload = {
    "name": "Тестовый пользователь",
    "login": "user1",
    "password": "12345"
}

resp = requests.post(
    f"{AUTH_URL}/register",
    headers=HEADERS_JSON,
    data=json.dumps(payload)
)

print("Register user")
print_response(resp)
```

```python
# Авторизация пользователя и подготовка headers

# POST /auth/token — логин
payload = {
    "username": "user1",
    "password": "12345"
}

resp = requests.post(f"{AUTH_URL}/token", data=payload)

print("Login")
print_response(resp)

token = resp.json().get("access_token")

HEADERS_AUTH = {
    "Authorization": f"Bearer {token}",
    "Content-Type": "application/json"
}
```

```python
# POST /notes — создать заметку

payload = {
    "title": "Моя первая заметка",
    "content": "Это пример текста заметки"
}

resp = requests.post(
    NOTES_URL,
    headers=HEADERS_AUTH,
    data=json.dumps(payload)
)

print("Create note")
print_response(resp)

note_id = resp.json().get("id")
```

```python
# GET /notes — получить список своих заметок
resp = requests.get(
    NOTES_URL,
    headers=HEADERS_AUTH
)
print("Get my notes")
print_response(resp)
```

```python
# GET /notes/{id}
resp = requests.get(
    f"{NOTES_URL}/{note_id}",
    headers=HEADERS_AUTH
)
print("Get note by id")
print_response(resp)
```

```python
# PATCH /notes/{id}
payload = {
    "title": "Обновлённый заголовок"
}
resp = requests.patch(
    f"{NOTES_URL}/{note_id}",
    headers=HEADERS_AUTH,
    data=json.dumps(payload)
)
print("Patch note")
print_response(resp)
```

```python
# PUT /notes/{id}
payload = {
    "title": "Новая версия заметки",
    "content": "Полностью заменённый текст"
}
resp = requests.put(
    f"{NOTES_URL}/{note_id}",
    headers=HEADERS_AUTH,
    data=json.dumps(payload)
)
print("Put note")
print_response(resp)
```

```python
# DELETE /notes/{id}
resp = requests.delete(
    f"{NOTES_URL}/{note_id}",
    headers=HEADERS_AUTH
)
print("Delete note")
print_response(resp)
```
